```python
import matplotlib.pyplot as plt
import numpy as np

# Data for population distribution
age_groups = ['0 to 20 Years', '21 to 64 Years', '65+ Years']
population_millions = [512, 807, 98]  # Population in millions for each group
percentages = [36.1, 57.0, 6.9]  # Percentages for each group
colors = ['yellow', 'blue', 'magenta']  # Colors matching the original chart

# Create a figure with 2 subplots (stacked area and bar chart)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Create stacked area chart on the first subplot (ax1)
ax1.fill_between([0, 20], 0, population_millions[0], color=colors[0], label=f'{age_groups[0]}: {population_millions[0]}M ({percentages[0]}%)')
ax1.fill_between([21, 64], 0, population_millions[1], color=colors[1], label=f'{age_groups[1]}: {population_millions[1]}M ({percentages[1]}%)')
ax1.fill_between([65, 100], 0, population_millions[2], color=colors[2], label=f'{age_groups[2]}: {population_millions[2]}M ({percentages[2]}%)')

# Set labels and title for the area chart
ax1.set_xlabel('Age')
ax1.set_ylabel('Population (in millions)')
ax1.set_title("India's Population Distribution by Age in 2022")
ax1.legend(loc='upper right')

# Adding text for population, median age, and total population
total_population = sum(population_millions)
ax1.text(50, 20, f'Population: {total_population} million\nMedian Age: 28 years', fontsize=12, ha='center')

# Create bar chart on the second subplot (ax2)
ax2.bar(age_groups, population_millions, color=colors)

# Set labels and title for the bar chart
ax2.set_xlabel('Age Groups')
ax2.set_ylabel('Population (in millions)')
ax2.set_title("Population Distribution by Age Groups")

# Display the plot
plt.tight_layout()
plt.show()
```
