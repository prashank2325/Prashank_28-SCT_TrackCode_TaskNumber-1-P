```python
import matplotlib.pyplot as plt
import numpy as np

# Data for population distribution
age_groups = ['0 to 20 Years', '21 to 64 Years', '65+ Years']
population_millions = [512, 807, 98]  # Population in millions for each group
percentages = [36.1, 57.0, 6.9]  # Percentages for each group
colors = ['yellow', 'blue', 'magenta']  # Colors matching the original chart

# Create the figure and axis
plt.figure(figsize=(8, 6))

# Create stacked area chart to represent population by age groups
plt.fill_between([0, 20], 0, population_millions[0], color=colors[0], label=f'{age_groups[0]}: {population_millions[0]}M ({percentages[0]}%)')
plt.fill_between([21, 64], 0, population_millions[1], color=colors[1], label=f'{age_groups[1]}: {population_millions[1]}M ({percentages[1]}%)')
plt.fill_between([65, 100], 0, population_millions[2], color=colors[2], label=f'{age_groups[2]}: {population_millions[2]}M ({percentages[2]}%)')

# Set labels, title, and legend
plt.xlabel('Age')
plt.ylabel('Population (in millions)')
plt.title("India's Population Distribution by Age in 2022")

# Adding text for population, median age, and total population
total_population = sum(population_millions)
plt.text(50, 20, f'Population: {total_population} billion\nMedian Age: 28 years', fontsize=12, ha='center')

# Displaying the legend
plt.legend(loc='upper right')

# Display the plot
plt.tight_layout()
plt.show()
```
